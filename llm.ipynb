```python
from transformers import AutoTokenizer, AutoModelForCausalLM
from transformers import pipeline

# 言語モデルのダウンロード
tokenizer = AutoTokenizer.from_pretrained("cerebras/Cerebras-GPT-590M")
model = AutoModelForCausalLM.from_pretrained("cerebras/Cerebras-GPT-590M")

# パイプラインの準備
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
# テキストの生成
text = "What is AI?"
generated_text = pipe(text, max_length=100, do_sample=False, no_repeat_ngram_size=2)[0]['generated_text']
print(generated_text)
```
